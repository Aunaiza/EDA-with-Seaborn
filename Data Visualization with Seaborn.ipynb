```python
  ###########################************************SCATTER PLOT ****************************##############################
    
%matplotlib inline    #magic function
import seaborn as sns
import matplotlib.pyplot as plt

height = [62, 64, 69, 75, 66,
          68, 65, 71, 76, 73]          # in inches
weight = [120, 136, 148, 175, 137,
          165, 154, 172, 200, 187]    # in pounds

# Do taller people tend to weigh more?

sns.scatterplot(x=height , y=weight)
plt.show()

# What do we conclude? A linear relationship exists between them if height grows weight also grows
```

```python
###########################************************COUNT PLOT ****************************##############################
    

gender = ["Female", "Female",
"Female", "Female",
"Male", "Male", "Male",
"Male", "Male", "Male"]

# How many samples belong to male and female?

sns.countplot(x=gender)
plt.show()

```

```python
#########################**********Adding 3rd variable with hue **********************##############################]
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
tips.head()





```

```python
########### Does high bill contribute to higher tips? #####################

sns.scatterplot(x="total_bill" , y="tip" , data=tips)
plt.show()
```

```python
###################***********How many of these data points are smokers? ******************#############

sns.scatterplot(x="total_bill" , y="tip" , data=tips ,hue="smoker")
plt.show()


```

```python
###################***********How many of these data points are male/female? ******************#############

sns.scatterplot(x="total_bill" , y="tip" , data=tips ,hue="sex",hue_order=["Female","Male"])
plt.show()


```

```python
##########################************** Customising Hue Colors ***************#####################

hue_colors = {"Female":"Red",
              "Male":"blue"}

sns.scatterplot(x="total_bill" , y="tip" , data=tips ,hue="sex",palette=hue_colors,hue_order=["Female","Male"])
plt.show()


```

```python
################################ Using hue with count plot ####################################

sns.countplot(x="sex",data =tips)
plt.show()

# around 160 males and 80 females

```

```python
sns.countplot(x="sex",data =tips , hue="smoker")
plt.show()

# out of 160 males 100 are non-smokers
```

# Joint Plot

```python
# 2 variables

sns.jointplot(data=df, x="tip", y="total_bill",kind="hex")

# hex ---> datapoints will be drawn as hexagons
# Where histo value for x and y are high we can see that the concentrauon of points is high there
# majority of the people have given tips between 2 to 4 $
# majority of the bill is b/w 10-20 $
```

```python
sns.jointplot(data=df, x="tip", y="total_bill",kind="reg")

# probability density function on histograms
# fits the regression line on data points (linear regression fit)
```
